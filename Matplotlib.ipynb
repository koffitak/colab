```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Les données à visualiser
prix_ms = [39, 50, 57, 69, 94, 100]
prix_apple = [93, 112, 104, 144, 170, 172]
annee = [2015, 2016, 2017, 2018, 2019, 2020]
```

```python
# visualiser les données
plt.plot(annee, prix_ms, label="Prix windows")
plt.plot(annee, prix_apple, label="Prix apple")
plt.legend()
plt.xlabel("Annee")
plt.ylabel("Prix des ordinateurs")
plt.title("Evolution des prix des ordinateurs durant les 5 dernières années")
```

```python
# visualiser les données
plt.scatter(annee, prix_ms, label="Prix windows", c='g')
plt.plot(annee, prix_apple, label="Prix apple", c='r')
plt.legend()
plt.xlabel("Annee")
plt.ylabel("Prix des ordinateurs")
plt.title("Evolution des prix des ordinateurs durant les 5 dernières années")
```

```python
# visualiser les données dans deux plans
plt.subplot(2,1,1)
plt.scatter(annee, prix_ms, label="Prix windows", c='lime')
plt.subplot(2,1,2)
plt.plot(annee, prix_apple, label="Prix apple", c='r')
```

```python
# visualiser les données dans deux plans
plt.subplot(1,2,1)
plt.scatter(annee, prix_ms, label="Prix windows", c='lime')
plt.subplot(1,2,2)
plt.plot(annee, prix_apple, label="Prix apple", c='r')
```

```python
# Visualiser les données en diagramme de bande
plt.bar(annee, prix_ms, label="Prix windows")
plt.bar(annee, prix_apple, label="Prix apple", bottom=prix_ms)
plt.legend()
plt.xlabel("Annee")
plt.ylabel("Prix des ordinateurs")
plt.title("Evolution des prix des ordinateurs durant les 5 dernières années")
```

```python
# Visualiser les données en diagramme de bande
index_annee = np.arange(len(annee))
a=0.25
plt.bar(index_annee, prix_ms, label="Prix windows", width=0.25)
plt.bar(index_annee+a, prix_apple, label="Prix apple", width=0.25)
plt.legend()
plt.xlabel("Annee")
plt.ylabel("Prix des ordinateurs")
plt.title("Evolution des prix des ordinateurs durant les 5 dernières années")
```
