List comprehension

```python
nombres = [14, 20, 30, 360, 25]
```

```python
nombres_paires = []
for i in nombres:
  if i%2==0:
    nombres_paires.append(i)
nombres_paires    
```

```python
nombres_paires = [x+2 for x in nombres if x%2==0]
nombres_paires
```

Map()  #Fait le lien entre une fonction et une itérable(list...)

```python
noms = ['amine','anouar','sama','omar']
```

```python
def verifie(x):
  if x[0]=='a':
    return x.upper()
```

```python
nom_a = list(map(verifie, noms))
nom_a
```

enumrate() 

```python
nombres = [12, 23, 20, 30, 35, 320, 469, 23, 56]
```

```python
for indice, valeur in enumerate(nombres):
  print(indice)
  print(valeur)
  print('_________')
```
